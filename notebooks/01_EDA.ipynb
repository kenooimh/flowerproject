```python
import torchvision
from torchvision import datasets
import matplotlib.pyplot as plt

# Download the Oxford 102 Flowers dataset
train_data = datasets.Flowers102(root='../data', split='train', download=True)
print(f"Dataset loaded. Total training images: {len(train_data)}")
```

```python
# Technical Metadata for the Oxford 102 Dataset
print(f"Number of classes: {len(train_data.classes)}")
print(f"Example Image Size: {train_data[0][0].size}") # Shows raw resolution
print(f"Dataset Split: Training (10 images/class), Validation (10 images/class), Test (Remainer)")
```

```python
# Extract labels and plot distribution
labels = train_data._labels
plt.figure(figsize=(12, 4))
plt.hist(labels, bins=102, color='teal', rwidth=0.8)
plt.title("Class Distribution (Oxford 102 Flowers)")
plt.xlabel("Class ID")
plt.ylabel("Count")
plt.show()
```

```python
# Show a few examples
fig, axes = plt.subplots(1, 5, figsize=(15, 5))
for i in range(5):
    img, label = train_data[i]
    axes[i].imshow(img)
    axes[i].set_title(f"Class: {label}")
    axes[i].axis('off')
plt.show()
```

Even distribution across flower class, all flowers class has 10 images across.
